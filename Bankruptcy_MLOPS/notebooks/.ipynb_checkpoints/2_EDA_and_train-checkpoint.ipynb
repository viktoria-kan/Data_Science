# Описание

Необходимо создать модель, которая будет прогнозировать банкроство компаний. Данные включают в себя финансовые показатели компаний. В данных не учитывается информация о сфере деятельности компании и инф-я о внешних факторах (например, экономическая обстановка в стране/мире, расположение, и др)

Банкротство компании — это неспособность предприятия в полном объеме удовлетворить требования кредиторов по денежным обязательствам и/или исполнить обязанности по уплате обязательных платежей.

Основные признаки:

**target** - целевая переменная: 1 - компания является банкротом; 0 - компания не является банкротом

**Применимо ко всем признакам:**

```python
def calculation_of_coeffs_div(data: pd.DataFrame, x: str, y: str, new_col: str,
                              perc: int) -> pd.DataFrame:
    """
    Расчёт финансовых коэффицицентов
    :param data: датасет
    :param x: признак, который будет в числителе   
    :param y: признак, который будет в знаменателе
    :param new_col: новый признак(коэффициент)
    :param perc: процент, на который нужно умножить результат
    :return: датасет
    """
    # Если в формуле не нужно умножать результат на 100, то параметр perc=1
    result = data[x] / data[y] * perc
    result[data[y] == 0] = 0
    data[new_col] = result
    return data
```

```python
def calculation_of_coeffs_sum(data: pd.DataFrame, x: str, y: str,
                              new_col: str) -> pd.DataFrame:
    """
    Расчёт финансовых коэффицицентов
    :param data: датасет
    :param x: первое слагаемое  
    :param y: второе слагаемое
    :param new_col: новый признак(коэффициент)
    :return: датасет
    """
    result = data[x] + data[y]
    data[new_col] = result
    return data
```

```python
def calculation_of_coeffs_subtraction(data: pd.DataFrame, x: str, y: str,
                                      new_col: str) -> pd.DataFrame:
    """
    Расчёт финансовых коэффицицентов
    :param data: датасет
    :param x: признак, из которого вычитаем (уменьшаемое)  
    :param y: признак, который вычитаем (вычитаемое)
    :param new_col: новый признак(коэффициент)
    :return: датасет
    """
    result = data[x] - data[y]
    data[new_col] = result
    return data
```

```python
def get_bins(data: Union[int, float],
             first_val: Union[int, float] = 0,
             second_val: Union[int, float] = 0) -> str:
    """
    Генерация бинов для разных признаков
    :param data: датасет
    :param first_val: первое пороговое значение 
    :param second_val: второе пороговое значение  
    :return: датасет
    """
    assert isinstance(data, (int, float)), "Неверный тип данных в признаке"
    result = ("Ниже нормы" if data < first_val else "Оптимально"
              if first_val <= data <= second_val else "Выше нормы")
    return result
```

```python
def barplot_group(data: pd.DataFrame, col_main: str, col_group: str,
                  title: str) -> None:
    """
    Отрисовка графика boxplot
    :param data: датасет
    :param col_main: признак для анализа по col_group
    :param col_group: признак для нормализации/группировки
    :param title: заголовок графика
    :return: график
    """

    plt.figure(figsize=(15, 6))

    # Создание копии данных для преобразования типа данных внутри функции
    data_copy = data[[col_main, col_group]].copy()
    data_copy[col_group] = data_copy[col_group].astype(str)

    grouped_data = (data_copy.groupby(
        [col_group])[col_main].value_counts(normalize=True).rename(
            'percentage').mul(100).reset_index().sort_values(col_group))

    ax = sns.barplot(x=col_main,
                     y="percentage",
                     hue=col_group,
                     data=grouped_data,
                     palette='rocket')

    for p in ax.patches:
        percentage = '{:.1f}%'.format(p.get_height())
        ax.annotate(
            percentage,
            (p.get_x() + p.get_width() / 2., p.get_height()),  # координата xy
            ha='center',  # центрирование
            va='center',
            xytext=(0, 7),
            textcoords='offset points',  # точка смещения относительно координаты
            fontsize=12)

    plt.title(title, fontsize=16)
    plt.ylabel('Percentage', fontsize=14)
    plt.xlabel(col_main, fontsize=14)
    plt.show()
```

```python
def features_selection(data: pd.DataFrame, data_type: str) -> pd.Index:
    """
    Определение признаков, соответствующих указанному типу данных
    :param data: датасет
    :param data_type: тип данных
    :return: список признаков
    """
    num_cols = data.select_dtypes(include=[data_type]).columns
    return num_cols
```

```python
def corr_filter(data: pd.DataFrame, indx: pd.Index,
                  bound: float) -> pd.Series:
    """
    Определение пар признаков, у которых корреляция выше заданного порога
    :param data: датасет
    :param indx: список признаков
    :param bound: значение порога корреляции
    :return: пары скоррелированных признаков
    """
    
    data_corr = data[indx].corr()
    corr_cols = data_corr[(abs(data_corr) >= bound) & (data_corr != 1)]
    pair_corr_cols = corr_cols.unstack().sort_values().drop_duplicates()
    
    return pair_corr_cols.dropna()
```

Для более тонкой настройки данных желательно вручную отсеивать такие признаки, учитывая их значение и влияние. Однако в данном проекте по максимуму всё автоматизируем

```python
def count_corr_features(high_corr_features: pd.Series) -> pd.Series:
    """
    Расчет количества скоррелированных признаков для каждого уникального признака
    :param high_corr_features: признаки с высокой корреляцией
    :return: количество скоррелированных признаков 
    """

    counts_corr = {}

    for index, corr_value in high_corr_features.items():

        features = index
        for feature in features:
            if feature not in counts_corr:
                counts_corr[feature] = 0
            counts_corr[feature] += 1

    corr_series = pd.Series(counts_corr, dtype=int).sort_values()
    return corr_series
```

```python
def drop_corr_features(data: pd.DataFrame, corr_features: pd.Series,
                       limit_bounds: int) -> pd.DataFrame:
    """
    Удаление признаков, у которых количество скоррелированных признаков 
    больше заданного значения
    :param data: датасет
    :param corr_features: кол-во коррелирующих признаков
    :param limit_bound: предельное значение коррелирующих признаков
    :return: датасет
    """

    columns_to_drop = corr_features[
        corr_features >= limit_bounds].index.tolist()

    data_filtered = data.drop(columns=columns_to_drop)
    return data_filtered
```

```python
def save_unique_train_data(data: pd.DataFrame,
                           unique_values_path: str,
                           target_column: Optional[str] = None) -> None:
    """
    Сохранение словаря с признаками и уникальными значениями
    :param drop_columns: список с признаками для удаления
    :param data: датасет
    :param target_column: целевая переменная
    :param unique_values_path: путь до файла со словарем
    :return: None
    """

    if target_column is not None:
        unique_df = data.drop(columns=[target_column], axis=1, errors="ignore")
    else:
        unique_df = data

    dict_unique = {
        key: unique_df[key].unique().tolist()
        for key in unique_df.columns
    }
    with open(unique_values_path, "w") as file:
        json.dump(dict_unique, file)
```

```python
def get_metrics(y_test: np.array, y_pred: np.array, y_proba: np.array) -> Dict:
    """Расчёт метрик для задачи классификации"""
    
    dict_metrics = {
        'roc_auc': round(roc_auc_score(y_test, y_proba[:, 1]), 3),
        'precision': round(precision_score(y_test, y_pred), 3),
        'recall': round(recall_score(y_test, y_pred), 3),
        'f1': round(f1_score(y_test, y_pred), 3),
        'logloss': round(log_loss(y_test, y_proba), 3)
    }

    return dict_metrics
```

```python
def check_overfitting(
        model: BaseEstimator, X_train: np.array, y_train: np.array,
        X_test: np.array, y_test: np.array,
        metric_func: Callable[[np.ndarray, np.ndarray], float]) -> None:
    """
    Проверка на переобучение для классификации
    """

    y_score_train = model.predict_proba(X_train)
    y_score_test = model.predict_proba(X_test)
    metric_train = metric_func(y_train, y_score_train[:, 1])
    metric_test = metric_func(y_test, y_score_test[:, 1])

    print(f'Train-Test delta = {(abs(metric_train - metric_test) / metric_test*100):.1f} %')
```

# Tuning (optuna)

```python
def objective(trial, X: np.array, y: np.array, N_FOLDS: int,
              RAND: int) -> np.array:
    """
    Целевая функция для поиска параметров
    :param trial: кол-во trials
    :param X: данные объект-признаки
    :param y: данные с целевой переменной
    :param N_FOLDS: кол-во фолдов
    :param RAND: random_state
    :return: среднее значение метрики по фолдам
    """

    penalty = trial.suggest_categorical('penalty',
                                        ['l1', 'l2', 'elasticnet', None])
    C = trial.suggest_loguniform('C', 1e-3, 1e3)
    solver = trial.suggest_categorical(
        'solver', ['liblinear', 'saga', 'lbfgs', 'newton-cg', 'sag'])
    tol = trial.suggest_loguniform('tol', 1e-6, 1e-2)
    class_weight = trial.suggest_categorical("class_weight", ['balanced'])
    random_state = trial.suggest_categorical("random_state", [RAND])

    l1_ratio = None
    if penalty == 'elasticnet':
        l1_ratio = trial.suggest_uniform('l1_ratio', 0, 1)

    # Исключение несовместимых комбинаций
    if penalty == 'l1' and solver not in ['liblinear', 'saga']:
        raise optuna.exceptions.TrialPruned()
    if penalty == 'l2' and solver == 'liblinear':
        raise optuna.exceptions.TrialPruned()
    if penalty == 'elasticnet' and solver != 'saga':
        raise optuna.exceptions.TrialPruned()
    if penalty == None and solver in ['liblinear', 'saga']:
        raise optuna.exceptions.TrialPruned()

    model = LogisticRegression(penalty=penalty,
                               C=C,
                               solver=solver,
                               tol=tol,
                               l1_ratio=l1_ratio,
                               class_weight=class_weight,
                               random_state=random_state)

    skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RAND)

    X_train_df = pd.DataFrame(X)
    aucs = []
    for train_index, valid_index in skf.split(X_train_df, y):
        X_train_fold, X_valid_fold = X_train_df.iloc[
            train_index], X_train_df.iloc[valid_index]
        y_train_fold, y_valid_fold = y.iloc[train_index], y.iloc[valid_index]

        model.fit(X_train_fold, y_train_fold)
        y_pred = model.predict_proba(X_valid_fold)[:, 1]
        auc = roc_auc_score(y_valid_fold, y_pred)
        aucs.append(auc)

    return np.mean(aucs)
```

```python
study_lg = optuna.create_study(direction="maximize")

def func(trial):
    return objective(trial, X_train_std, y_train, training['n_folds'],
                     training['random_state'])


optuna.logging.set_verbosity(optuna.logging.WARNING)
study_lg.optimize(func, n_trials=training['n_trials'], show_progress_bar=True)
```

```python
y_pred = model.predict(X_test_std)
y_score = model.predict_proba(X_test_std)

# Определение индексов TN и FN, тк вкладываемся именно в них
tn_indices = np.where((y_test == 0) & (y_pred == 0))[0]
fn_indices = np.where((y_test == 1) & (y_pred == 0))[0]

# Создание DataFrame только для компаний из TP и FN
invest_companies = np.concatenate((tn_indices, fn_indices))
invest_probabilities = y_score[invest_companies]

investment_df = pd.DataFrame({
    'Company_Index': invest_companies,
    'Bankruptcy_Probability': invest_probabilities
})

# Общий бюджет инвестирования
total_investment = 100000
```

```python
def allocate_investments(df: pd.DataFrame,
                         total_investment: int) -> pd.DataFrame:
    """Распределение инвестиций с учетом вероятности банкротства"""

    df['Share_invest'] = 1 - df['Bankruptcy_Probability']
    sum_share_invest = df['Share_invest'].sum()
    df['Amount_invest'] = (df['Share_invest'] /
                           sum_share_invest) * total_investment
    return df
```

```python
# Распределение инвестиций
invest_results = allocate_investments(investment_df, total_investment)

print("Вероятности банкротства и распределение инвестиций:")
print(invest_results[['Company_Index', 'Bankruptcy_Probability', 'Amount_invest']])
```

```python
def finance_bankruptcy(model: BaseEstimator, y_test: np.array,
                       X_test: np.array, invest_cost: [int, float],
                       invest_rate: float) -> int:
    """Расчет конечной прибыли/убытков"""

    tn, fp, fn, tp = confusion_matrix(y_test, model.predict(X_test)).ravel()

    # Расходы на инвестиции в компании, которые не обанкротились (TN)
    good_invest_costs = tn * invest_cost

    # Потери от вложений в компании, которые обанкротились (FN)
    losses_invest = fn * invest_cost

    # Прибыль от удачных вложений (TN) с учетом доходности
    profit_good_investment = tn * invest_cost * invest_rate

    # Общая выгода
    total_benefit = profit_good_investment - losses_invest - good_invest_costs

    return total_benefit
```

```python
# Распределение инвестиций
benefit = finance_bankruptcy(model, y_test, X_test_std, invest_cost,
                                    invest_rate)

print("Итоговая прибыль/убытки")
print(benefit)
```
