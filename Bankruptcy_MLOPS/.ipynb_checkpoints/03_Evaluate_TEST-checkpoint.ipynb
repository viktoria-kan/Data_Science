```python
def check_columns_evaluate(data: pd.DataFrame,
                           unique_values_path: str,
                           flg_evaluate: bool) -> pd.DataFrame:
    with open(unique_values_path, encoding='utf-8') as json_file:
        unique_values = json.load(json_file)

    column_sequence = list(unique_values.keys())

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ target –≤ column_sequence, –µ—Å–ª–∏ —ç—Ç–æ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    if not flg_evaluate and 'target' not in column_sequence:
        column_sequence.append('target')

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    for col in column_sequence:
        if col not in data.columns:
            data[col] = 0

    # –£–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫—Ä–æ–º–µ target, –µ—Å–ª–∏ —ç—Ç–æ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    if flg_evaluate:
        data = data[[col for col in column_sequence if col in data.columns]]
#         print('with drop target')
    else:
        data = data[[col for col in column_sequence if col in data.columns or col == 'target']]
#         print('without drop target')

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Å—Ç–æ–ª–±—Ü–∞ target –≤ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
#     if not flg_evaluate and 'target' not in data.columns:
#         print('Error: target column is missing in the training data.')

    return data[column_sequence]
```

```python
def objective(trial, data_x, data_y, n_folds, random_state):
    """
    –¶–µ–ª–µ–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    :param trial: –∫–æ–ª-–≤–æ trials
    :param data_x: –¥–∞–Ω–Ω—ã–µ –æ–±—ä–µ–∫—Ç-–ø—Ä–∏–∑–Ω–∞–∫–∏
    :param data_y: –¥–∞–Ω–Ω—ã–µ —Å —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
    :param n_folds: –∫–æ–ª-–≤–æ —Ñ–æ–ª–¥–æ–≤
    :param random_state: random_state
    :return: —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –ø–æ —Ñ–æ–ª–¥–∞–º
    """

    penalty = trial.suggest_categorical('penalty',
                                        ['l1', 'l2', 'elasticnet', None])
    C = trial.suggest_loguniform('C', 1e-3, 1e3)
    solver = trial.suggest_categorical(
        'solver', ['liblinear', 'saga', 'lbfgs', 'newton-cg', 'sag'])
    tol = trial.suggest_loguniform('tol', 1e-6, 1e-2)
    class_weight = trial.suggest_categorical("class_weight", ['balanced'])
    random_state = trial.suggest_categorical("random_state", [random_state])

    l1_ratio = None
    if penalty == 'elasticnet':
        l1_ratio = trial.suggest_uniform('l1_ratio', 0, 1)

    # –ò—Å–∫–ª—é—á–µ–Ω–∏–µ –Ω–µ—Å–æ–≤–º–µ—Å—Ç–∏–º—ã—Ö –∫–æ–º–±–∏–Ω–∞—Ü–∏–π
    if penalty == 'l1' and solver not in ['liblinear', 'saga']:
        raise optuna.exceptions.TrialPruned()
    if penalty == 'l2' and solver == 'liblinear':
        raise optuna.exceptions.TrialPruned()
    if penalty == 'elasticnet' and solver != 'saga':
        raise optuna.exceptions.TrialPruned()
    # if penalty == 'none' and solver in ['liblinear', 'saga']:
    if penalty == None and solver in ['liblinear', 'saga']:
        raise optuna.exceptions.TrialPruned()

    skf = StratifiedKFold(n_splits=n_folds,
                          shuffle=True,
                          random_state=random_state)
    cv_predicts = np.empty(n_folds)

    x_train_df = pd.DataFrame(data_x)

    # aucs = []
    # for train_index, valid_index in skf.split(x_train_df, data_y):
    for idx, (train_idx, test_idx) in enumerate(skf.split(data_x, data_y)):
        x_train_fold, x_valid_fold = x_train_df.iloc[
            train_idx], x_train_df.iloc[test_idx]
        y_train_fold, y_valid_fold = data_y.iloc[train_idx], data_y.iloc[
            test_idx]

        model = LogisticRegression(penalty=penalty,
                                   C=C,
                                   solver=solver,
                                   tol=tol,
                                   l1_ratio=l1_ratio,
                                   class_weight=class_weight,
                                   random_state=random_state)

        model.fit(
            x_train_fold,
            y_train_fold,
        )
        predict = model.predict_proba(x_valid_fold)[:, 1]
        cv_predicts[idx] = roc_auc_score(y_valid_fold, predict)
        # auc = roc_auc_score(y_valid_fold, predict)
        # aucs.append(auc)

    # return np.mean(aucs)
    return np.mean(cv_predicts)


def find_optimal_params(data_train: pd.DataFrame, data_test: pd.DataFrame,
                        **kwargs) -> Study:
    """
    –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∏ –º–æ–¥–µ–ª–∏
    :param data_train: –¥–∞—Ç–∞—Å–µ—Ç train
    :param data_test: –¥–∞—Ç–∞—Å–µ—Ç test
    :return: [LGBMClassifier tuning, Study]
    """
    x_train, x_test, y_train, y_test = get_train_test_data(
        data_train=data_train,
        data_test=data_test,
        target=kwargs["target_column"])

    study = optuna.create_study(direction="maximize", study_name="LogReg")

    def function(trial):
        return objective(trial, x_train, y_train, kwargs["n_folds"],
                         kwargs["random_state"])

    study.optimize(function,
                   n_trials=kwargs["n_trials"],
                   show_progress_bar=True)
    return study


def train_model(
        data_train: pd.DataFrame,
        data_test: pd.DataFrame,
        study: Study,
        target: str,
        # metric_path: str,
        **kwargs) -> LogisticRegression:

    # get data
    x_train_std, x_test_std, y_train, y_test = get_train_test_data(
        data_train=data_train, data_test=data_test, target=target)

        
    finish_test_preds = []
    finish_test_preds_proba = []

    cv = StratifiedKFold(n_splits=kwargs['n_folds'],
                         shuffle=True,
                         random_state=kwargs['random_state'])
    cv_predicts_val = np.empty(kwargs['n_folds'])

#     x_train_df = pd.DataFrame(data_train)

    for idx, (train_idx, test_idx) in enumerate(cv.split(x_train_std, y_train)):
        x_train_, x_val = x_train_std.iloc[train_idx], x_train_std.iloc[test_idx]
        y_train_, y_val = y_train.iloc[train_idx], y_train.iloc[test_idx]

        clf = LogisticRegression(**study.best_params)
        clf.fit(x_train_, y_train_)

        # OOF
        preds_val_proba = clf.predict_proba(x_val)[:, 1]
        cv_predicts_val[idx] = roc_auc_score(y_val, preds_val_proba)

        # # holdout
        # preds_test = clf.predict(x_test_std)
        # preds_test_proba = clf.predict_proba(x_test_std)

        # holdout
        preds_test = clf.predict(x_test_std)
        preds_test_proba = clf.predict_proba(x_test_std)

        finish_test_preds.append(preds_test)
        finish_test_preds_proba.append(preds_test_proba)

        # print(f"id = {idx}", cv_predicts_val[idx], '\n')

        # print(np.mean(cv_predicts_val))

        # save metrics
        #     metrics = save_metrics(data_x=x_test,
        #                            data_y=y_test,
        #                            model=clf,
        #                            metric_path=kwargs['metrics_path'])

        metrics = save_metrics(data_x=x_test_std,
                               data_y=y_test,
                               model=clf,
                               metric_path=kwargs['metrics_path'])
        
    return clf, metrics


def save_metrics(data_x: pd.DataFrame, data_y: pd.Series, model: object,
                 metric_path: str) -> None:

    result_metrics = create_dict_metrics(
        y_test=data_y,
        y_predict=model.predict(data_x),
        y_probability=model.predict_proba(data_x),
    )
    return result_metrics


#     with open(metric_path, "w") as file:
#         json.dump(result_metrics, file)


def create_dict_metrics(y_test: pd.Series, y_predict: pd.Series,
                        y_probability: pd.Series) -> dict:

    dict_metrics = {
        "roc_auc": round(roc_auc_score(y_test, y_probability[:, 1]), 3),
        "precision": round(precision_score(y_test, y_predict), 3),
        "recall": round(recall_score(y_test, y_predict), 3),
        "f1": round(f1_score(y_test, y_predict), 3),
        "logloss": round(log_loss(y_test, y_probability), 3),
    }
    return dict_metrics
```

```python
def save_unique_train_data(data: pd.DataFrame,
                           unique_values_path: str,
                           target_column: Optional[str] = None) -> None:
    """
    –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è —Å –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
    :param drop_columns: —Å–ø–∏—Å–æ–∫ —Å –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param target_column: —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è
    :param unique_values_path: –ø—É—Ç—å –¥–æ —Ñ–∞–π–ª–∞ —Å–æ —Å–ª–æ–≤–∞—Ä–µ–º
    :return: None
    """

    if target_column is not None:
        unique_df = data.drop(columns=[target_column], axis=1, errors="ignore")
    else:
        unique_df = data

    dict_unique = {
        key: unique_df[key].unique().tolist()
        for key in unique_df.columns
    }
    with open(unique_values_path, "w") as file:
        json.dump(dict_unique, file)
```

```python
#–°–û–ó–î–ê–ù–ò–ï –§–ê–ô–õ–ê feature.json
def generate_features_json(uniq_values_path, output_path):
    with open(uniq_values_path, 'r', encoding='utf-8') as f:
        uniq_values = json.load(f)

    features = {}
    for key, values in uniq_values.items():
        if values:
            first_value = values[0]
            if isinstance(first_value, int):
                feature_type = "int"
            elif isinstance(first_value, float):
                feature_type = "float"
            elif isinstance(first_value, str):
                feature_type = "str"
            else:
                feature_type = "unknown"
            
            features[key] = feature_type

    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(features, f, indent=4)
```

```python
def calculation_of_coeffs_div(data: pd.DataFrame, x: str, y: str, new_col: str,
                              perc: int) -> pd.DataFrame:
    """
    –†–∞—Å—á—ë—Ç —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏—Ü–µ–Ω—Ç–æ–≤
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param x: –ø—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –≤ —á–∏—Å–ª–∏—Ç–µ–ª–µ   
    :param y: –ø—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –≤ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª–µ
    :param new_col: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫(–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç)
    :param perc: –ø—Ä–æ—Ü–µ–Ω—Ç, –Ω–∞ –∫–æ—Ç–æ—Ä—ã–π –Ω—É–∂–Ω–æ —É–º–Ω–æ–∂–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """
    # –ï—Å–ª–∏ –≤ —Ñ–æ—Ä–º—É–ª–µ –Ω–µ –Ω—É–∂–Ω–æ —É–º–Ω–æ–∂–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ 100, —Ç–æ –ø–∞—Ä–∞–º–µ—Ç—Ä perc=1 
    result = data[x] / data[y] * perc
    result[data[y] == 0] = 0
    data[new_col] = result
    return data


def calculation_of_coeffs_sum(data: pd.DataFrame, x: str, y: str,
                              new_col: str) -> pd.DataFrame:
    """
    –†–∞—Å—á—ë—Ç —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏—Ü–µ–Ω—Ç–æ–≤
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param x: –ø–µ—Ä–≤–æ–µ —Å–ª–∞–≥–∞–µ–º–æ–µ  
    :param y: –≤—Ç–æ—Ä–æ–µ —Å–ª–∞–≥–∞–µ–º–æ–µ
    :param new_col: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫(–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç)
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """
    result = data[x] + data[y]
    data[new_col] = result
    return data


def calculation_of_coeffs_subtraction(data: pd.DataFrame, x: str, y: str,
                                      new_col: str) -> pd.DataFrame:
    """
    –†–∞—Å—á—ë—Ç —Ñ–∏–Ω–∞–Ω—Å–æ–≤—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏—Ü–µ–Ω—Ç–æ–≤
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param x: –ø—Ä–∏–∑–Ω–∞–∫, –∏–∑ –∫–æ—Ç–æ—Ä–æ–≥–æ –≤—ã—á–∏—Ç–∞–µ–º (—É–º–µ–Ω—å—à–∞–µ–º–æ–µ)  
    :param y: –ø—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π –≤—ã—á–∏—Ç–∞–µ–º (–≤—ã—á–∏—Ç–∞–µ–º–æ–µ)
    :param new_col: –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫(–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç)
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """
    result = data[x] - data[y]
    data[new_col] = result
    return data
```

```python
def execute_tasks_from_file(data: pd.DataFrame,
                            file_path: str) -> pd.DataFrame:
    """
    –†–∞—Å—á—ë—Ç –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ –æ—Å–Ω–æ–≤–µ JSON —Ñ–∞–π–ª–∞ —Å –∑–∞–¥–∞—á–∞–º–∏
    :param data: –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
    :param file_path: –ø—É—Ç—å –∫ JSON —Ñ–∞–π–ª—É —Å –∑–∞–¥–∞—á–∞–º–∏
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        tasks = json.load(file)

    # –°–ª–æ–≤–∞—Ä—å –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å—Ç—Ä–æ–∫ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏–∏
    functions = {
        'calculation_of_coeffs_div': calculation_of_coeffs_div,
        'calculation_of_coeffs_sum': calculation_of_coeffs_sum,
        'calculation_of_coeffs_subtraction': calculation_of_coeffs_subtraction
    }

    # –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∑–∞–¥–∞—á
    for task in tasks:
        func = functions[task["function"]]
        args = task["args"]
        data = func(data, *args)

    return data
```

```python
def get_bins(data: Union[int, float],
             first_val: Union[int, float] = 0,
             second_val: Union[int, float] = 0) -> str:
    """
    –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –±–∏–Ω–æ–≤ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param first_val: –ø–µ—Ä–≤–æ–µ –ø–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 
    :param second_val: –≤—Ç–æ—Ä–æ–µ –ø–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ  
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """
    assert isinstance(data, (int, float)), "–ù–µ–≤–µ—Ä–Ω—ã–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –≤ –ø—Ä–∏–∑–Ω–∞–∫–µ"
    result = ("–ù–∏–∂–µ –Ω–æ—Ä–º—ã" if data < first_val else "–û–ø—Ç–∏–º–∞–ª—å–Ω–æ"
              if first_val <= data <= second_val else "–í—ã—à–µ –Ω–æ—Ä–º—ã")
    return result


def features_selection(data: pd.DataFrame, data_type: str) -> pd.Index:
    """
    –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —É–∫–∞–∑–∞–Ω–Ω–æ–º—É —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param data_type: —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö
    :return: —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    """
    num_cols = data.select_dtypes(include=[data_type]).columns
    return num_cols


def check_columns_evaluate(data: pd.DataFrame,
                           unique_values_path: str) -> pd.DataFrame:
    """
    –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ train –∏ —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ train
    :param data: –¥–∞—Ç–∞—Å–µ—Ç test
    :param unique_values_path: –ø—É—Ç—å –¥–æ —Å–ø–∏—Å–∫–∞ —Å –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ train 
    :return: –¥–∞—Ç–∞—Å–µ—Ç test
    """
    with open(unique_values_path, encoding='utf-8') as json_file:
        unique_values = json.load(json_file)

    column_sequence = unique_values.keys()

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    for col in column_sequence:
        if col not in data.columns:
            data[col] = 0

    # –£–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    data = data[[col for col in column_sequence if col in data.columns]]

    return data[column_sequence]
```

```python
def check_columns_evaluate(data: pd.DataFrame,
                           unique_values_path: str,
                           flg_evaluate: bool) -> pd.DataFrame:
    with open(unique_values_path, encoding='utf-8') as json_file:
        unique_values = json.load(json_file)

    column_sequence = unique_values.keys()

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    for col in column_sequence:
        if col not in data.columns:
            data[col] = 0

    # –£–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫—Ä–æ–º–µ target, –µ—Å–ª–∏ —ç—Ç–æ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    if flg_evaluate:
        data = data[[col for col in column_sequence if col in data.columns]]
    else:
        data = data[[col for col in column_sequence if col in data.columns or col == 'target']]

    return data[column_sequence]
```

```python
def pipeline_preprocess(data: pd.DataFrame,
                        flg_evaluate: bool = True,
                        **kwargs):
    """
    –ü–∞–π–ø–ª–∞–π–Ω –ø–æ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–µ –¥–∞–Ω–Ω—ã—Ö
    :param data: –¥–∞—Ç–∞—Å–µ—Ç
    :param flg_evaluate: —Ñ–ª–∞–≥ –¥–ª—è evaluate
    :return: –¥–∞—Ç–∞—Å–µ—Ç
    """

    data = execute_tasks_from_file(data, kwargs['tasks_path'])

    if flg_evaluate:
        data = check_columns_evaluate(
            data=data, unique_values_path=kwargs["unique_values_path"])
    else:
        save_unique_train_data(
            data=data,
            drop_columns=kwargs["drop_columns"],
            #             target_column=kwargs["target_column"],
            target_column=None,
            unique_values_path=kwargs["unique_values_path"],
        )

    # getbins
    assert isinstance(
        kwargs["bins_columns"],
        dict), "–ü–æ–¥–∞–π—Ç–µ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –±–∏–Ω–∞—Ä–∏–∑–∞—Ü–∏–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ dict"
    # bins
    for key in kwargs["bins_columns"].keys():
        data[f"{key}_bins"] = data[key].apply(lambda x: get_bins(
            x,
            first_val=kwargs["bins_columns"][key][0],
            second_val=kwargs["bins_columns"][key][1],
        ))

    num_selected_cols = features_selection(data, preproc['data_type'][0])
    cat_selected_cols = features_selection(data, preproc['data_type'][1])
      
    data = pd.get_dummies(data,
                          columns=cat_selected_cols,
                          drop_first=True,
                          dtype=int)
    
    scaler = StandardScaler()
    data[num_selected_cols] = scaler.fit_transform(data[num_selected_cols])

    data = check_columns_evaluate(
        data=data, unique_values_path=kwargs["uniq_val_path_with_binar"])

    return data
```
