## Удаление дублирующихся признаков

Репетиторы записывают одни и те же услуги в разных формулировках, поэтому некоторые услуги дублируются, объединим их:

```python
# ВАРИАНТ 2 для объединения столбцов по значению в новый
#Функция для создания нового столбца на основе других
def create_and_remove_columns(df, columns_to_check, new_column_name,
                              condition_value):
    # Создайте новый столбец, применяя условие
    df[new_column_name] = (df[columns_to_check] == condition_value).any(
        axis=1).astype(int)

    # Удалите старые столбцы
    df.drop(columns=columns_to_check, axis=1, inplace=True)
```

```python
create_and_remove_columns(df, [
    'CELTA Certificate'
], 'имеет_серт_CELTA', 1)
```

```python
create_and_remove_columns(df, [
    'сертификат TESOL', 'сертификат Teaching English for Speakers of Other Languages',
    'TESOL Certificate'
], 'имеет_серт_TESOL', 1)
```

Тк есть признаки бакалавр, магистр, аспирантура, которые по отдельности несут мало информации, можно создать новые признаки на их основе: 

```python
def new_feature_graduate(row):
    job_list = []
    if any(row.iloc[[19,20,21]] == 1) and all(
            row.iloc[i] == 0 for i in [22, 23, 45]):
        job_list.append('Иное высшее образование')
    elif all(row.iloc[[19, 20, 21, 22, 23, 45]]==0):
        job_list.append('Образование не указано')
    else:
#         job_list.append('Высшее языковое образование')
        job_list.append('Высшее языковое образование')
    return ', '.join(job_list)
```

Тк у некоторых репетиторов нет фиксированной цены, рассчитаем среднее значение

```python
def clean_and_average(price):
#     price = re.sub(r'[^\d\s-]', '', price)  # Удаляем все символы, кроме цифр, пробелов и тире
    prices = [int(p) for p in re.findall(r'\d+', price)]  # Извлекаем все числа из строки
    if prices:
        return np.mean(prices)  # Возвращаем среднее арифметическое
    return np.nan
#     print(prices)
# Применяем функцию к столбцу 'price' и создаем новый столбец 'mean_price'
# df2['mean_price'] = df2['price'].apply(clean_and_average)
df['mean_price'] = df['price'].apply(clean_and_average)
```

```python
# Функция для очистки и вычисления среднего duration
def clean_and_average_dur(duration):
#     price = re.sub(r'[^\d\s-]', '', price)  # Удаляем все символы, кроме цифр, пробелов и тире
    clean_duration = [int(p) for p in re.findall(r'\d+', duration)]  # Извлекаем все числа из строки
    if clean_duration:
        return np.mean(clean_duration)  # Возвращаем среднее арифметическое
    else:
        return 
    return np.nan
#     print(prices)
# Применяем функцию к столбцу 'duration' и создаем новый столбец 'mean_duration'
df['mean_duration'] = df['duration'].apply(clean_and_average_dur)
```

```python
# ВАРИАНТ 2
# Бины переименовать
def get_bins_duration(data: int) -> str:
    """
    Генерация бинов для признака "duration"
    """
    if data <= 45:
        return '0-45'
    elif 30 < data < 90:
        return '30-90'
    elif data >= 90:
        return '90-240'
    else:
        return 'по запросу'
#     return None
df['duration_bins'] = df.mean_duration.transform(lambda x: get_bins_duration(x))
```

```python
# Объединим некоторые значения признака "passport" 
def get_bins_passport(data: int) -> str:
    """
    Генерация бинов для признака "passport"
    """
#     if isinstance(data, (int, float)):
    if str(data) == 'Данные компании проверены':
        return 'Паспорт проверен'
    elif str(data) == 'Данные компании не проверены':
        return 'Паспорт не проверен'
    elif str(data) == 'Мало отзывов по вашей услуге' or str(data) =='Нет отзывов по вашей услуге':
        return 'Документы не предоставлены'
    else:
        return data 
    return None
```

```python
def barplot_group(col_main: str, col_group: str, title: str) -> None:
    """
    Построение barplot с нормированными данными с выводом значений на графике
    """
    
    plt.figure(figsize=(15, 6))

    data = (df.groupby([col_group])[col_main]
            .value_counts(normalize=True)
            .rename('percentage')
            .mul(100)
            .reset_index()
            .sort_values(col_group))

    ax = sns.barplot(x=col_main, y="percentage", hue=col_group, data=data, palette='rocket')

    for p in ax.patches:
        percentage = '{:.1f}%'.format(p.get_height())
        ax.annotate(percentage,  # текст
                    (p.get_x() + p.get_width() / 2.,
                     p.get_height()),  # координата xy
                    ha='center',  # центрирование
                    va='center',
                    xytext=(0, 7),
                    textcoords='offset points',  # точка смещения относительно координаты
                    fontsize=12)

    plt.title(title, fontsize=16)
    plt.ylabel('Percentage', fontsize=14)
    plt.xlabel(col_main, fontsize=14)
    plt.show()
```

```python
(df.groupby(['gender'])['teacher_educ']
            .value_counts(normalize=True)
            .rename('percentage')
            .mul(100)
            .reset_index()
            .sort_values("teacher_educ"))
```

```python
barplot_group("teacher_educ", "gender", 'Education-gender')
```
