```python
#ф-я 3 для объединения столбцов по значению в новый, остановилась на ней.
# Определите функцию для создания нового столбца и удаления старых
def create_and_remove_columns(df, columns_to_check, new_column_name, condition_value):
    # Создайте новый столбец, применяя условие
    df[new_column_name] = (df[columns_to_check] == condition_value).any(axis=1).astype(int)
    
    # Удалите старые столбцы
    df.drop(columns=columns_to_check, axis=1, inplace=True)
```

```python
def splitter(names):
    list_first_names = []
    for i in names:
        first_name = i.split(' ')[0]
        list_first_names.append(first_name)
    return list_first_names
```

```python
# Объединим некоторые значения признака "passport" 
def get_bins_age(data: int) -> str:
    """
    Генерация бинов для признака "passport"
    """
#     if isinstance(data, (int, float)):
    if str(data) == 'Данные компании проверены':
        return 'Паспорт проверен'
    elif str(data) == 'Данные компании не проверены':
        return 'Паспорт не проверен'
    elif str(data) == 'Мало отзывов по вашей услуге' or str(data) =='Нет отзывов по вашей услуге':
        return 'Документы не предоставлены'
    else:
        return data
    return None
```

```python
# Функция для очистки и вычисления среднего
def clean_and_average(price):
#     price = re.sub(r'[^\d\s-]', '', price)  # Удаляем все символы, кроме цифр, пробелов и тире
    prices = [int(p) for p in re.findall(r'\d+', price)]  # Извлекаем все числа из строки
    if prices:
        return np.mean(prices)  # Возвращаем среднее арифметическое
    return np.nan
#     print(prices)
# Применяем функцию к столбцу 'price' и создаем новый столбец 'mean_price'
# df2['mean_price'] = df2['price'].apply(clean_and_average)
df['mean_price'] = df['price'].apply(clean_and_average)
```

```python
def barplot_group(col_main: str, col_group: str, title: str) -> None:
    """
    Построение barplot с нормированными данными с выводом значений на графике
    """
    
    plt.figure(figsize=(15, 6))

    data = (df.groupby([col_group])[col_main]
            .value_counts(normalize=True)
            .rename('percentage')
            .mul(100)
            .reset_index()
            .sort_values(col_group))

    ax = sns.barplot(x=col_main, y="percentage", hue=col_group, data=data, palette='rocket')

    for p in ax.patches:
        percentage = '{:.1f}%'.format(p.get_height())
        ax.annotate(percentage,  # текст
                    (p.get_x() + p.get_width() / 2.,
                     p.get_height()),  # координата xy
                    ha='center',  # центрирование
                    va='center',
                    xytext=(0, 7),
                    textcoords='offset points',  # точка смещения относительно координаты
                    fontsize=12)

    plt.title(title, fontsize=16)
    plt.ylabel('Percentage', fontsize=14)
    plt.xlabel(col_main, fontsize=14)
    plt.show()
```

```python
(df.groupby(['gender'])['teacher_educ']
            .value_counts(normalize=True)
            .rename('percentage')
            .mul(100)
            .reset_index()
            .sort_values("teacher_educ"))
```

```python
barplot_group("teacher_educ", "gender", 'Education-gender')
```

```python
def get_bins_mark(data: int) -> str:
    """
    Генерация бинов для признака "mark"
    """
    #     if isinstance(data, (int, float)):
    if data <= 45:
        return '45<='
    elif 45 < data < 90:
        return '45-90'
    else:
        return '90>='
#     return None
```

```python
df['duration_bins'] = df.mean_duration.transform(lambda x: get_bins_duration(x))
```

```python
def get_bins_reviews(data: int) -> str:
    """
    Генерация бинов для признака "reviews"
    """
    #     if isinstance(data, (int, float)):
    if data < 50:
        return 'Менее 50 отзывов'
    elif 50 <= data <= 250:
        return '50-250 отзывов'
    else:
        return 'Более 250 отзывов'
#     return None
```

```python
df['reviews_bins'] = df.mean_duration.transform(lambda x: get_bins_duration(x))
```

```python
def get_bins_stars(row: int) -> str:
    """
    Генерация бинов для признака "stars"
    """
    #     if isinstance(data, (int, float)):
    if row['5_stars_percent'] >= 95:
        return 'отзывов на 5 звезд больше 95%'
    elif any(row[['1_stars_percent', '2_stars_percent']])!=0:
        return 'есть плохие отзывы'
    else:
        return 0
df['отзывы'] = df.apply(get_bins_stars, axis=1)
```

```python
df = pd.get_dummies(df, columns=['отзывы'], drop_first=True)
```

```python
def get_bins_price(data: int) -> str:
    """
    Генерация бинов для признака "mean_price"
    """
    if data < 1500:
        return 'Средняя цена <1500'
    elif 1500 <= data <= 3500:
        return 'Средняя цена 1500-3500'
    else:
        return 'Средняя цена >3500'
#     return None
# df2['ср_цена'] = df2.apply(get_bins_price, axis=1)
```

```python
df['средняя_цена'] = df.mean_price.transform(lambda x: get_bins_price(x))
```
