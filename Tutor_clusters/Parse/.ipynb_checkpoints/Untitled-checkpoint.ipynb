```python
# Сбор данных с каждой карточки репетитора
# С кнопкой Показать еще 20

def call_func_text(repetitors, max_objects=60):
    rate_list = []
    basic_inf_list = []
    services_list = []
    counter = 0
#     max_objects = 60
    driver = webdriver.Chrome()
#     driver.get(URL)

    for rep in repetitors:
        first_func = tutor_card(rep, driver)

        second_func = rate_info(first_func)
        rate_list.append(second_func)

        third_func = basic_info(first_func)
        basic_inf_list.append(third_func)

        fourth_func = search_services(first_func, driver)
        services_list.append(fourth_func)
        
        counter += 1
        
    # Если собрано 20 элементов и счетчик меньше max_elements, нажимаем кнопку "Показать ещё 20"
        if counter % 20 == 0 and counter < max_objects:
            # Нажимаем кнопку "Показать ещё 20"
            driver.get(URL)
#             button = driver.find_element("xpath", '//*[@id="page"]/div/main/div/div/div[2]/div[2]/div/button')
            button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//*[@id='page']/div/main/div/div/div[2]/div[2]/div/button[contains(text(), 'Показать ещё 20')]")))
            button.click()
            print('КНОПКА НАЖАТА')
            # Ожидаем обновления страницы
            WebDriverWait(driver, 5).until(EC.staleness_of(button))

        # Если счетчик достиг максимального количества элементов, выходим из цикла
        if counter >= max_objects:
            print('КОнец')
            break
        
    driver.quit()
#     return rate_list, basic_inf_list, services_list
    return {
        'rate_list': rate_list,
        'basic_inf_list': basic_inf_list,
        'services_list': services_list
    }


def tutor_card(rep, driver):
    names_url = rep.find(
        'a', {'class': 'ui_1hi7c ui_RF7aD ui_3j-OD'}).get("href")
    url = "https://profi.ru{}".format(names_url)
#     driver = webdriver.Chrome()
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    return soup


def rate_info(soup):
    all_rates = soup.find_all('div', {'class': '_lWApzmR'})
    counts_marks = []
    for i in all_rates:
        i = str(i).replace('<div class="_lWApzmR">', '').replace('</div>', '')
        counts_marks.append(i)
    return counts_marks


def basic_info(soup):
    list_of_inform = []
    all_info = soup.find('div', {'data-shmid': 'profileOIO'}).text
    edit_info = process_text(all_info)
    edit_info = re.sub(" +", " ", edit_info)
    list_of_inform.append(edit_info)
    return list_of_inform


def process_text(text):
    result = []
    i = 0
    while i < len(text):
        if text[i].isupper() and (i == 0 or not re.match(r'[A-ZА-Я\d\s]', text[i - 1])):
            result.append(' ')
        if not re.match(r'\w', text[i]):
            result.extend([text[i], ' '])
        else:
            result.append(text[i])
        i += 1
    return ''.join(result)
```

```python
# Сбор данных с каждой карточки репетитора 
# РАБОТАЕТ, но нет кнопки Показать еще 20

def internal_data_collection(rep, driver):
    rate_list = []
    basic_inf_list=[]
    services_list=[]
#     driver = webdriver.Chrome()
#     driver.get(URL)
#     for rep in repetitors:
    first_func = tutor_card(rep, driver)
        
    second_func = rate_info(first_func)
    rate_list.append(second_func)
        
    third_func = basic_info(first_func)
    basic_inf_list.append(third_func)
        
    fourth_func = search_services(first_func, driver)
    services_list.append(fourth_func)
        
    driver.quit()   
#     return rate_list, basic_inf_list, services_list
    return {
        'rate_list': rate_list,
        'basic_inf_list': basic_inf_list,
        'services_list': services_list
    }

def tutor_card(rep, driver):
    names_url = rep.find('a', {'class':'ui_1hi7c ui_RF7aD ui_3j-OD'}).get("href")
    url = "https://profi.ru{}".format(names_url)
    driver = webdriver.Chrome()
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    return soup

def rate_info(soup):
    all_rates = soup.find_all('div', {'class':'_lWApzmR'})
    counts_marks=[]
    for i in all_rates:
        i = str(i).replace('<div class="_lWApzmR">','').replace('</div>','')
        counts_marks.append(i)
    return counts_marks 

def basic_info(soup):
    list_of_inform=[]
    all_info = soup.find('div', {'data-shmid':'profileOIO'}).text
    edit_info=process_text(all_info)
    edit_info = re.sub(" +", " ", edit_info)
    list_of_inform.append(edit_info)
    return list_of_inform

def process_text(text):
    result = []
    i = 0
    while i < len(text):
        if text[i].isupper() and (i == 0 or not re.match(r'[A-ZА-Я\d\s]', text[i - 1])):
            result.append(' ')
        if not re.match(r'\w', text[i]):
            result.extend([text[i], ' '])
        else:
            result.append(text[i])
        i += 1
    return ''.join(result)
```

```python
# вариант N
# РАБОТАЕТ
def search_services(soup, driver):
    all_service = []
    amount = soup.find('span', {'class': 'ui-text _3xKhc83 _3QMH9Cr'})
    if amount == None:
        serv1 = soup.find('table', {'class': 'price-list desktop-profile__prices'})
        serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
        serv_final = edit_text(serv2)           
        all_service.extend(serv_final)
    else:
        try:
            clicker = driver.find_element(
                    "link text", f"Все услуги и цены {amount.text}").click()
            html2 = driver.page_source
            soup2 = BeautifulSoup(html2, 'lxml')
            serv1 = soup2.find_all(
                    'table', {'class': 'price-list desktop-profile__prices'})[1]
            serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
            serv_final = edit_text(serv2)         
            all_service.extend(serv_final)
        except:
            all_service.extend('ERROR')
    return all_service
  
def edit_text(serv2):
    servs = []
    for j in serv2:
        j = str(j).replace(
            '<td class="item_name item_name-bold"><span>',
            '').replace('</span><br/></td>', '').replace(
                '</span><br/><span class="ui-text _3eH689t _3QMH9Cr">', '')
        servs.append(j)
    return servs
```
