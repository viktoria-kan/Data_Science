```python
# Всё слеплено в кучу
def external_data_collection(repetitors):
    list_of_names = []
    list_of_marks = []
    list_of_commend = []
    confirm_list = []
    list_of_attest = []
    list_of_discounts = []
    list_of_prices = []
    duration_list = []
    for i in repetitors:
        name = i.find('a', {'class': 'ui_1hi7c ui_RF7aD ui_3j-OD'}).text

        mark = i.find('span', {
            'data-shmid': 'profileIndicators_ratingBlock'
        }).text

        commend = i.find(
            'div', {
                'class':
                'ui_2MYaG ui_2SXfw ui_RF7aD ui_3Y-uv ui_Rff0c ui_9cYgw ui_VLOD7'
            }).text.split('отзыв')[1]
        commend = re.sub(r"\bа|\bов", "", commend)
        if commend == '':
            commend = 'Статуса нет'
            list_of_commend.append(commend)
        else:
            list_of_commend.append(commend)

        passport = i.find('span', {'data-shmid': 'profile_pinf'}).text

        attestat = i.find(
            'p', {'class': 'ui_2MYaG ui_2SXfw ui_RF7aD ui_1hoX9 _3P8_yr8'})
        if attestat == None:
            attestat = 0
            list_of_attest.append(attestat)
        else:
            list_of_attest.append(attestat.text)

        discount = i.find('div', {'data-shmid': 'profile_discount-block'})
        if discount == None:
            discount = 0
            list_of_discounts.append(discount)
        else:
            list_of_discounts.append('yes')

        price = i.find('td', {'class': 'item_value'}).text.split('₽')[0]

        duration = i.find(
            'span', {
                'class': 'ui-text price-list__duration _3xKhc83 _5DgZvLV'
            }).text.split('— ')[1]

        list_of_names.append(name)
    return list_of_names
```

```python
# Сбор данных с каждой карточки репетитора 
# С кнопкой Показать еще 20

def call_func_text(repetitors):
    rate_list = []
    basic_inf_list=[]
    services_list=[]
    counter = 0
    max_objects = 5000
    driver = webdriver.Chrome()
    
    while counter < max_objects:
        try:
            # Нажимаем на кнопку "Показать ещё 20"
            show_more_button = driver.find_element_by_css_selector(
                '.show-more-button')
            show_more_button.click()
        except NoSuchElementException:
            # Если кнопка "Показать ещё 20" не найдена, прекращаем цикл
            break

        # Ожидание загрузки новых объектов (может потребоваться настроить время ожидания)
        time.sleep(2)
    
    for rep in repetitors:
        first_func = tutor_card(rep, driver)
        
        second_func = rate_info(first_func)
        rate_list.append(second_func)
        
        third_func = basic_info(first_func)
        basic_inf_list.append(third_func)
        
        fourth_func = search_services(first_func, driver)
        services_list.append(fourth_func)
        
    driver.quit()   
#     return rate_list, basic_inf_list, services_list
    return {
        'rate_list': rate_list,
        'basic_inf_list': basic_inf_list,
        'services_list': services_list
    }

def tutor_card(rep, driver):
    names_url = rep.find('a', {'class':'ui_1hi7c ui_RF7aD ui_3j-OD'}).get("href")
    url = "https://profi.ru{}".format(names_url)
#     driver = webdriver.Chrome()
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    return soup

def rate_info(soup):
    all_rates = soup.find_all('div', {'class':'_lWApzmR'})
    counts_marks=[]
    for i in all_rates:
        i = str(i).replace('<div class="_lWApzmR">','').replace('</div>','')
        counts_marks.append(i)
    return counts_marks 

def basic_info(soup):
    list_of_inform=[]
    all_info = soup.find('div', {'data-shmid':'profileOIO'}).text
    edit_info=process_text(all_info)
    edit_info = re.sub(" +", " ", edit_info)
    list_of_inform.append(edit_info)
    return list_of_inform

def process_text(text):
    result = []
    i = 0
    while i < len(text):
        if text[i].isupper() and (i == 0 or not re.match(r'[A-ZА-Я\d\s]', text[i - 1])):
            result.append(' ')
        if not re.match(r'\w', text[i]):
            result.extend([text[i], ' '])
        else:
            result.append(text[i])
        i += 1
    return ''.join(result)
```

```python
# вариант N
# РАБОТАЕТ
def search_services(soup, driver):
    all_service = []
    amount = soup.find('span', {'class': 'ui-text _3xKhc83 _3QMH9Cr'})
    if amount == None:
        serv1 = soup.find('table', {'class': 'price-list desktop-profile__prices'})
        serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
        serv_final = edit_text(serv2)           
        all_service.extend(serv_final)
    else:
        clicker = driver.find_element(
                "link text", f"Все услуги и цены {amount.text}").click()
        html2 = driver.page_source
        soup2 = BeautifulSoup(html2, 'lxml')
        serv1 = soup2.find_all(
                'table', {'class': 'price-list desktop-profile__prices'})[1]
        serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
        serv_final = edit_text(serv2)         
        all_service.extend(serv_final)
    return all_service
  
def edit_text(serv2):
    servs = []
    for j in serv2:
        j = str(j).replace(
            '<td class="item_name item_name-bold"><span>',
            '').replace('</span><br/></td>', '').replace(
                '</span><br/><span class="ui-text _3eH689t _3QMH9Cr">', '')
        servs.append(j)
    return servs
```

```python
def called_func(repetitors, max_objects=60):
    counter = 0
    ext_list = []
    inter_list = []

    options = Options()
    options.add_argument('--disable-blink-features=AutomationControlled')
    driver = webdriver.Chrome(options=options)
    driver.get(URL)

    # Нажимаем кнопку "Показать ещё 20" если это необходимо
    def click_show_more():
        button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//*[@id='page']/div/main/div/div/div[2]/div[2]/div/button[contains(text(), 'Показать ещё 20')]")))
        button.click()
        print('КНОПКА НАЖАТА')
        WebDriverWait(driver, 5).until(EC.staleness_of(button))

    for rep in repetitors:
        external_data = external_data_collection(rep)
        ext_list.append(external_data)

        internal_data = call_func_text(rep, driver)
        inter_list.append(internal_data)

        counter += 1

        if counter % 20 == 0 and counter < max_objects:
            click_show_more()

        if counter >= max_objects:
            print('КОнец')
            break

    driver.quit()
    return ext_list, inter_list
```
