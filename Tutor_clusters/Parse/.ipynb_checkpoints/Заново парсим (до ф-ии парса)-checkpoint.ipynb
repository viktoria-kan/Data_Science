```python
import requests
from bs4 import BeautifulSoup
import threading

url = "https://фиас.онлайн/aef2c735-2e21-42e5-8638-9667bade6a2b/"
base_url = "https://фиас.онлайн"

response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")

streets = soup.find_all("div", class_="col-md-3")

streets_houses_dict = {}


def get_houses(street_name, street_url):
    response = requests.get(street_url)
    soup = BeautifulSoup(response.text, "lxml")

    houses = soup.find_all("table")[1].find_all("td")[1].find_all("a")
    houses_list = []

    for i in houses:
        houses_list.append(i.string)

    streets_houses_dict[street_name] = houses_list


def async_get_houses():
    threads_list = []

    for street in streets:
        street_name = street.find("a").text.split(" ")[0]
        street_url = street.find("a").get("href")
        get_street_houses = threading.Thread(
            target=get_houses, args=(street_name, base_url + street_url)
        )
        get_street_houses.start()

        threads_list.append(get_street_houses)

    for t in threads_list:
        t.join()


async_get_houses()
```
