```python
# Создаем семафор с ограничением в 3 потока
semaphore = threading.Semaphore(3)

def parse_element(element, queue):
    rate_list = []
    names_url = element.find('a', {'class': 'ui_1hi7c ui_RF7aD ui_3j-OD'}).get("href")
    url = "https://profi.ru{}".format(names_url)
    
    # Захватываем семафор перед выполнением задачи
    with semaphore:
        driver = webdriver.Chrome()
        driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        counts = soup.find_all('div', {'class': '_lWApzmR'})
        counts = edit_text3(counts)
        rate_list.append(counts)
        driver.quit()
    
    # Помещаем результат в очередь
    queue.put(rate_list)

def edit_text3(counts):
    counts_marks=[]
    for i in counts:
        inner_list = []
        for j in i:
            j = str(j).replace('<div class="_lWApzmR">','').replace('</div>','')
            inner_list.append(j)
        counts_marks.append(inner_list)
    return counts_marks

def rate_info(reps):
    queue = multiprocessing.Queue()  # Создаем очередь для хранения результатов
    
    threads = []
    for element in reps:
        # Создаем поток для обработки элемента и передаем очередь
        thread = threading.Thread(target=parse_element, args=(element, queue))
        thread.start()
        threads.append(thread)
    
    # Ожидаем завершения всех потоков
    for thread in threads:
        thread.join()
    
    # Извлекаем результаты из очереди
    rate_list = []
    while not queue.empty():
        rate_list.extend(queue.get())
    
    return rate_list

# Пример вызова функции
reps = soup.find_all('div', {'class': 'seamless-profile desktop-profile'})  # Замените 'some_class' на ваш класс элементов
result = rate_info(reps)
print(result)

# Пример вызова функции
# reps = soup.find_all('div', {'class': 'seamless-profile desktop-profile'}) 
```

```python
#Инф-я об образовании, опыте и доп инф-я
def basic_info(repetitors):
    list_of_inform=[]
    for i in repetitors:
        names_url = i.find('a', {'class':'ui_1hi7c ui_RF7aD ui_3j-OD'}).get("href")
        url = "https://profi.ru{}".format(names_url)
        driver = webdriver.Chrome()
        driver.get(url)
        response = requests.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html,'lxml')
        all_info = soup.find('div', {'data-shmid':'profileOIO'}).text
        edit_info=process_text(all_info)
        edit_info = re.sub(" +", " ", edit_info)
        list_of_inform.append(edit_info)
        print(f'____\n{edit_info}')
#         print(ahaha)
    return list_of_inform

# Функция для обработки текста
def process_text(text):
    # Инициализируем переменную для результата
    result = []
    # Проходим по каждому символу в тексте
    i = 0
    while i < len(text):
        # Если текущий символ - заглавная буква и это первая буква или перед ней не буква или цифра, добавляем пробел
        if text[i].isupper() and (i == 0 or not re.match(r'[A-ZА-Я\d\s]', text[i - 1])):
            result.append(' ')
        # Если текущий символ - не буква и не цифра, добавляем пробел перед и после
        if not re.match(r'\w', text[i]):
#             result.extend([' ', text[i], ' '])
            result.extend([text[i], ' '])
        else:
            # Иначе, добавляем текущий символ к результату
            result.append(text[i])
        i += 1
    # Объединяем результат в строку и возвращаем  
    return ''.join(result)
```
