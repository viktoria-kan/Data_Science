```python
def condition_one(status_text):
    if len(status_text) >= 2:
        second_status_element = status_text[1]
        status_text = second_status_element.get_text(strip=True)
        return status_text
    else:
        return "Статуса нет"

def condition_two(information):
    return 'Нет' if information is None else information

def external_data_collection(i):

    tutor_data = {}

    tutor_data["name"] = i.find('h1', {'class': 'ui_2MYaG ui_2SXfw ui_RF7aD ui_3owa1'}).text
    tutor_data["mark"] = i.find('span', {'data-shmid': 'profileIndicators_ratingBlock'}).text

    status_text = i.find_all('div', {'class': 'ui_jKnqN'})  
    status_text = condition_one(status_text)
    tutor_data["status"] = status_text
        
    tutor_data["passport"] = i.find('span', {'data-shmid':'profile_pinf'}).text
        
    attestation = i.find('span', {'class':'ui-text sft-info-title _1JCcbzP _1i2s3kL _2iyzK60 _3YppjWm'})
    attestation = condition_two(attestation)
    tutor_data["attestation"] = attestation

    discount = i.find('div', {'data-shmid':'profile_discount-block'})
    discount = condition_two (discount)
    tutor_data["discount"] = discount
        
    tutor_data["price"] = i.find('td', {'class':'item_value'}).text.split('₽')[0]
        
#     duration = i.find('span', {'class':'ui-text price-list__duration _3xKhc83 _5DgZvLV'}).text.split('— ')[1]
    duration = i.find('span', {'class':'ui-text price-list__duration _3xKhc83 _5DgZvLV'})
    duration = condition_two(duration)    
    tutor_data["duration"] = duration
        
    location = i.find_all('div', {'class':'ui-text _1Jx20En _3xKhc83 _2iyzK60 _3YppjWm'})
    location = condition_two(location)
    tutor_data["location"] = location

    return tutor_data
```

```python
def search_services(soup, driver):
    serv2='Нет'
    amount = soup.find('span', {'class': 'ui-text _3xKhc83 _3QMH9Cr'})
    if amount == None:
        serv1 = soup.find('table', {'class': 'price-list desktop-profile__prices'})
        serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
        serv_final = edit_text(serv2)           
    else:
        try:           
            time.sleep(2)
            button = WebDriverWait(driver, 3).until(
            EC.visibility_of_element_located((
                By.XPATH,
                "//*[@id='prices']/div/div/a/span"
            )))
            button.click()
            WebDriverWait(driver, 3).until(EC.staleness_of(button))
            html2 = driver.page_source
            soup2 = BeautifulSoup(html2, 'lxml')
            serv1 = soup2.find_all(
                    'table', {'class': 'price-list desktop-profile__prices'})[1]
            serv2 = serv1.find_all('td', {'class': 'item_name item_name-bold'})
            serv_final = edit_text(serv2)         
        except:
            return serv2 

    return serv_final
  
def edit_text(serv2):
    servs = []
    for j in serv2:
        j = str(j).replace(
            '<td class="item_name item_name-bold"><span>',
            '').replace('</span><br/></td>', '').replace(
                '</span><br/><span class="ui-text _3eH689t _3QMH9Cr">', '')
        servs.append(j)
    return servs
```

```python
# ЭТО ДЛЯ НОВОГО DF
df2 = pd.DataFrame(columns=['5 stars', '4 stars', '3 stars', '2 stars', '1 stars'])

# Проходим по каждому элементу списка и заменяем значения
for index, item in enumerate(variable1):
    rate_list = item['rate_list']
    rate_list = [int(div.get_text(strip=True)) for div in rate_list]  # Преобразование в целые числа
    df2.loc[index] = rate_list
```

```python
# ЭТО ДЛЯ УЖЕ СОЗДАННОГО DF
# Добавьте пустые колонки для каждого элемента rate_list
for i in range(5):  # Здесь 5 - количество элементов в rate_list
    df1[f'{i + 1}_stars'] = None

# Пройдитесь по вашим данным и добавьте значения в колонки
for i, item in enumerate(variable1):
    for j, value in enumerate(item['rate_list']):
        df1.at[i, f'{5 - j}_stars'] = int(value.text)
```

```python
for col in df1['attestation']:
    df1['attestation'] = df1['attestation'].apply(lambda x: x.text if x != 'Нет' else x)
```

```python
for col in df1['duration']:
    df1['duration'] = df1['duration'].apply(lambda x: x.split(' мин')[0] if x != 'Нет' else x)
```

```python
def process_location(locations):
    return [location.text if location != 'Нет' else location for location in locations]

# Применить функцию к каждому элементу в столбце 'location'
df1['location'] = df1['location'].apply(process_location)
```

```python
df['location'] = df['location'].apply(process_location)
```

```python
# Создать новые столбцы с нулями
df1['Teaches_at_home'] = 0
df1['Teaches_at_clients_place'] = 0
df1['Teaches_remotely'] = 0

# Функция для установки значений в новых столбцах
def set_values(row):
    if 'Принимает у себя' in row['location']:
        row['Teaches_at_home'] = 1
    if 'Выезд к клиенту' in row['location']:
        row['Teaches_at_clients_place'] = 1
    if 'Работает дистанционно' in row['location']:
        row['Teaches_remotely'] = 1
    return row

# Применить функцию к каждой строке
df1 = df1.apply(set_values, axis=1)
```

### Ищем ключевые слова

```python
# Поиск ключ слов для basic info
def check_keywords(list_of_dicts, keywords):
    results = {keyword: [] for keyword in keywords}

    for text_dict in list_of_dicts:
        text = text_dict['basic_inf_list']  # Получаем текст из словаря
        keyword_found = {keyword: 0 for keyword in keywords}
        
        for keyword in keywords:
            if keyword.lower() in text.lower():
                keyword_found[keyword] = 1

        for keyword, found in keyword_found.items():
            results[keyword].append(found)

    return results

def load_keywords_from_file(file_path):
    keywords = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            keyword = line.strip()
            if keyword:
                keywords.append(keyword)

    return keywords

# Укажите путь к вашему файлу с ключевыми словами
file_path = 'keywords_for_basic_inf.txt'

# Загрузите ключевые слова из файла в список
input_keywords2 = load_keywords_from_file(file_path)

# Теперь input_keywords2 содержит ключевые слова из файла
# print(input_keywords2)

results = check_keywords(variable2, input_keywords2)

# for keyword, presence_list in results.items():
#     print(f"{keyword}: {presence_list}")
```

```python
new_columns = list(results.keys())  # Получаем названия столбцов

# Добавляем новые столбцы в существующий DataFrame
for col_name, col_data in results.items():
    df1[col_name] = col_data
```

```python
# Поиск ключ слов для services 
def check_keywords(list_of_dicts, keywords):
    results = {keyword: [] for keyword in keywords}

    for text_dict in list_of_dicts:
        text_list = text_dict['services_list']  # Получаем список строк из словаря
        keyword_found = {keyword: 0 for keyword in keywords}
        
        for keyword in keywords:
            for text in text_list:
                if keyword.lower() in text.lower():
                    keyword_found[keyword] = 1
                    break  # Если нашли хотя бы один раз, переходим к следующему ключевому слову

        for keyword, found in keyword_found.items():
            results[keyword].append(found)

    return results

def load_keywords_from_file(file_path):
    keywords = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            keyword = line.strip()
            if keyword:
                keywords.append(keyword)

    return keywords

# Укажите путь к вашему файлу с ключевыми словами
file_path = 'keywords_for_services.txt'

# Загрузите ключевые слова из файла в список
input_keywords2 = load_keywords_from_file(file_path)

# Теперь input_keywords2 содержит ключевые слова из файла
# print(input_keywords2)

results = check_keywords(variable3, input_keywords2)

# for keyword, presence_list in results.items():
#     print(f"{keyword}: {presence_list}")
```
