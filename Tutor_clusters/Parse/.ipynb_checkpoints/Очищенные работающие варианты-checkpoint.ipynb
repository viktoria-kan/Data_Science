```python
def add_spaces_before_capital_letters(text):
#     pattern = r'(?<=[a-zа-яё])(?=[A-ZА-ЯЁ](?![A-ZА-ЯЁa-zа-яё]))'
    pattern = r'(?<=[^ ])(?=[A-ZА-ЯЁ](?![A-ZА-ЯЁ]))(?![A-ZА-ЯЁ])'
    result = re.sub(pattern, ' ', text)
    return result
```

```python
def add_spaces(text):
    result = ""
    for i, char in enumerate(text):
        if i == 0 or i == len(text) - 1 or char.isupper():
            result += " "
        result += char
    return result.strip()

text = "ПриветМирЕГЭ"
result = add_spaces(text)
print(result)
```

```python
def add_spaces(text):
    result = ""
    for i, char in enumerate(text):
        if i > 0 and char.isupper() and not text[i-1].isspace():
            result += " "
        result += char
    return result.strip()

text = "ПриветМирЕГЭ"
result = add_spaces(text)
print(result)
```
